Part - A : Implement the MLP using a simple data (2 inputs and binary output - AND, OR and XOR)

1.: Implementing the MLP for AND GATE                                 


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Define the dataset for AND gate
X_train = np.array([[0, 0],
                   [0, 1],
                   [1, 0],
                   [1, 1]])

y_train = np.array([0, 0, 0, 1])

# Define the MLP model with a simpler architecture
model = Sequential()

# Input layer with 2 nodes and a hidden layer with 2 neurons (simpler model for AND gate)
model.add(Dense(2, input_dim=2, activation='relu'))  # Hidden layer with ReLU activation
model.add(Dense(1, activation='sigmoid'))  # Output layer with sigmoid activation for binary output

# Compile the model with a smaller learning rate
optimizer = Adam(learning_rate=0.05)  # Decrease learning rate
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Train the model for more epochs (increased to 1000)
model.fit(X_train, y_train, epochs=50, batch_size=4, verbose=1)

# Evaluate the model
_, accuracy = model.evaluate(X_train, y_train)
print(f"Model Accuracy: {accuracy*100:.2f}%")

# Predict outputs using the trained model
predictions = model.predict(X_train)

# Apply thresholding (e.g., output > 0.5 is considered '1')
predicted_output = (predictions >= 0.5).astype(int)

# Print the predicted output
print("\nPredicted Outputs:")
for i, pred in enumerate(predicted_output):
    print(f"Input: {X_train[i]}, Predicted Output: {pred[0]}")
```

2.: Implementing the MLP for OR GATE





```python

# Define the dataset for OR gate
X_train_or = np.array([[0, 0],
                       [0, 1],
                       [1, 0],
                       [1, 1]])

y_train_or = np.array([0, 1, 1, 1])  # OR gate output

# Define the MLP model for OR gate
model_or = Sequential()
model_or.add(Dense(2, input_dim=2, activation='relu'))  # Hidden layer with 2 neurons
model_or.add(Dense(1, activation='sigmoid'))  # Output layer with sigmoid activation

# Compile the model
optimizer_or = Adam(learning_rate=0.05)
model_or.compile(optimizer=optimizer_or, loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model_or.fit(X_train_or, y_train_or, epochs=50, batch_size=4, verbose=1)

# Evaluate the model
_, accuracy_or = model_or.evaluate(X_train_or, y_train_or)
print(f"OR Gate Model Accuracy: {accuracy_or*100:.2f}%")

# Predict outputs
predictions_or = model_or.predict(X_train_or)
predicted_output_or = (predictions_or >= 0.5).astype(int)

# Print the predicted outputs
print("\nPredicted Outputs for OR Gate:")
for i, pred in enumerate(predicted_output_or):
    print(f"Input: {X_train_or[i]}, Predicted Output: {pred[0]}")
```

3.: Implementing the MLP for XOR GATE


```python


# Define the dataset for XOR gate
X_train_xor = np.array([[0, 0],
                        [0, 1],
                        [1, 0],
                        [1, 1]])

y_train_xor = np.array([0, 1, 1, 0])  # XOR gate output

# Define the MLP model for XOR gate
model_xor = Sequential()

# Increase complexity: Add more neurons to the hidden layer
model_xor.add(Dense(8, input_dim=2, activation='tanh'))  # Hidden layer with 8 neurons and tanh activation
model_xor.add(Dense(1, activation='sigmoid'))  # Output layer with sigmoid activation

# Compile the model with a smaller learning rate
optimizer_xor = Adam(learning_rate=0.001)  # Reduce learning rate
model_xor.compile(optimizer=optimizer_xor, loss='binary_crossentropy', metrics=['accuracy'])

# Train the model for more epochs
model_xor.fit(X_train_xor, y_train_xor, epochs=300, batch_size=4, verbose=1)  # Batch size = 4 to match the number of samples

# Evaluate the model
_, accuracy_xor = model_xor.evaluate(X_train_xor, y_train_xor)
print(f"XOR Gate Model Accuracy: {accuracy_xor*100:.2f}%")

# Predict outputs
predictions_xor = model_xor.predict(X_train_xor)
predicted_output_xor = (predictions_xor >= 0.5).astype(int)

# Print the predicted outputs
print("\nPredicted Outputs for XOR Gate:")
for i, pred in enumerate(predicted_output_xor):
    print(f"Input: {X_train_xor[i]}, Predicted Output: {pred[0]}")
```
